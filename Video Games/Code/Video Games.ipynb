```python
list_of_publishers=['Electronic Arts','Activision','Ubisoft','Nintendo','Thq','Sony Computer Entertainment'
                    ,'Take-Two Interactive','Sega','Konami Digital Entertainment','Namco Bandai Games'
                    ,'Microsoft Game Studios','Atari'
                    ,'Warner Bros. Interactive Entertainment'
                    ,'Capcom','Square Enix','Tecmo Koei','Midway Games','Vivendi Games','Eidos Interactive'
                    ,'Codemasters','Disney Interactive Studios']
```

```python
def publisher_set(x):
    if x in list_of_publishers:
        return x
    else:
        return 'Others'
```

```python
train['PUBLISHER']=train['PUBLISHER'].apply(publisher_set)
```

```python
test['PUBLISHER']=test['PUBLISHER'].apply(publisher_set)
```

```python
dummies=['PUBLISHER','CATEGORY','CONSOLE']
label_en=['RATING']
```

```python
train1=train.copy()
```

```python
train1=pd.get_dummies(train,columns=dummies,drop_first=True)
```

```python
test1=pd.get_dummies(test,columns=dummies,drop_first=True)
```

```python
for feature in label_en:    
    labels_ordered=train1.groupby(train1[feature])['SalesInMillions'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    train1[feature]=train1[feature].map(labels_ordered)
    test1[feature]=test1[feature].map(labels_ordered)
    
```

```python
test1['RATING']=test1['RATING'].apply(lambda x: 6.0 if pd.isnull(x) else x)
```

```python
scale_columns=['CRITICS_POINTS','USER_POINTS','YEAR']
```

```python
X=train1.drop('SalesInMillions',axis=1)
y=train1['SalesInMillions']
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scale=MinMaxScaler()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
X_train[scale_columns]=scale.fit_transform(X_train[scale_columns])
X_test[scale_columns]=scale.transform(X_test[scale_columns])
test1[scale_columns]=scale.transform(test1[scale_columns])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
```

```python
rf=RandomForestRegressor(bootstrap=False, max_depth=100, max_features='sqrt',
                      min_samples_leaf=1, min_samples_split=10,
                      n_estimators=2000,random_state=101)
```

```python
rf.fit(X_train,y_train)
```

```python
pred_rf=rf.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,pred_rf)**0.5
```

```python
from xgboost import XGBRegressor
```

```python
xgb = XGBRegressor(learning_rate =0.06, n_estimators=3461, 
                                     max_depth=3,min_child_weight=0 ,
                                     gamma=0, subsample=0.7,
                                     colsample_bytree=0.7,
                                     objective= 'reg:squarederror',nthread=4,
                                     scale_pos_weight=1,seed=1337, 
                                     reg_alpha=0.00006,
                                     random_state=101
                )
```

```python
xgb.fit(X_train,y_train)
```

```python
xgb_pred=xgb.predict(X_test)
```

```python
mean_squared_error(y_test,xgb_pred)**0.5
```
